```python
import pandas as pd
import nltk
from nltk.stem import WordNetLemmatizer
from nltk.stem import PorterStemmer
nltk.download('punkt')
nltk.download('wordnet')
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
df = pd.read_csv('/content/spotify_millsongdata.csv' , encoding = 'utf-8')
```

```python
df = df.drop('link' , axis = 1).reset_index(drop=True)
```

```python
df = df.sample(5000).reset_index()
```

```python
df.drop('index' , axis = 1)
```

```python
df['text'][0]
```

```python
df['text'] = df['text'].str.lower().replace(r'^\W\s' , " ").replace(r'\n' , " ", regex = True)
```

```python
lemmatizer = WordNetLemmatizer()
```

```python
def tokenLemma(txt):
  token = nltk.word_tokenize(txt)
  a = [lemmatizer.lemmatize(w) for w in token]
  return " ".join(a)
```

```python
stemmer = PorterStemmer()
def tokenStemmer(txt):
  token = nltk.word_tokenize(txt)
  a = [stemmer.stem(w) for w in token ]
  return " ".join(a)
```

```python
tokenStemmer('you are beautiful')
```

```python
tokenLemma('you are beautiful')
```

```python
df['text'] = df['text'].apply(lambda x: tokenLemma(x))
```

```python
tfidf = TfidfVectorizer(analyzer = 'word' , stop_words = 'english')
matrix = tfidf.fit_transform(df['text'])
```

```python
similarity = cosine_similarity(matrix)
```

```python
def recommender(song_df):
    idx = df[df['song'] == song_df].index[0]
    distances = sorted(list(enumerate(similarity[idx])),reverse=True,key=lambda x:x[1])

    songs = []
    for s_id in distances[1:5]:
        songs.append(df.iloc[s_id[0]].song)

    return songs
```

```python
recommender("Call Me Irresponsible")
```
